To predict the percentage of an student based on the no. of study hours

## import Python Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## import dataset

```python
data=pd.read_csv("data.csv")
print("Imported data successfull")
```

```python
data.head(10)
```

# #plot the graph for detail analysis of data set

```python
data.plot(x='Hours',y='Scores',style='1')
plt.title('hours vs percentage')
plt.xlabel('hours studied')
plt.ylabel('percentage scored')
plt.show()
```

```python
data.plot.pie(x='Hours',y='Scores')
```

# #by analysing the data we came to the conclusion that as study hours increases the students percentage also increases 

```python
X= data.iloc[:,:-1].values
Y=data.iloc[:,1].values
```

# #training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test ,Y_train, Y_test=train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
print("training complete")
```

```python
print(X_test)
print("prediction of score")
Y_pred=regressor.predict(X_test)
print(Y_pred)
```

# #checking the accuracy

```python
df=pd.DataFrame({'actual':Y_test,'predicted':Y_pred})
df
```

# #predicting with custom input

```python
hours=[[9.25]]
pred=regressor.predict(hours)
print(pred)
```

# #evaluating the model

```python
from sklearn import metrics
print('Mean absolute error',
      metrics.mean_absolute_error(Y_test,Y_pred))
```
